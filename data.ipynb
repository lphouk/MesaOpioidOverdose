```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from IPython.display import display
```

```python
def AgeGroupMerge(age):
    if age in ['0-14', '15-19']:
        return 'Teen'
    elif age in ['20-24', '25-29', '30-34', '35-39']:
        return 'Young Adult'
    elif age in ['40-44', '45-49', '50-54', '55-59', '60-64']:
        return 'Middle-aged Adult'
    else:
        return 'Elderly'
```

```python
def weekend(day):
    if day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']:
        return 0
    else:
        return 1
```

```python
data = pd.read_csv('data.csv')
```

```python
#data['opioidOverdose'] = [int(i) for i in data['opioidOverdose']]
data['opioidOverdose'] = data['opioidOverdose'].astype(int)


data['pdNarcan'] = pd.to_numeric(data['pdNarcan'], errors='coerce')
data = data.dropna(subset=['pdNarcan'])
data['pdNarcan'] = data['pdNarcan'].astype(int)

data['pdSave'] = pd.to_numeric(data['pdSave'], errors='coerce')
data = data.dropna(subset=['pdSave'])
data['pdSave'] = data['pdSave'].astype(int)

data['initialTreatment'] = data['initialTreatment'].fillna('None')
data['naloxone'] = data['initialTreatment'].apply(lambda x: 1 if 'Naloxone' in str(x) else 0)

data['fatal'] = data['finalCallStatus'].apply(lambda x: 1 if 'Patient' in str(x) else 0)

data = data[data['opioidOverdose'] != 0]

data = data.drop('compCode', axis=1)

data['gender'] = data['gender'].fillna('NA')
data = data[(data['gender'] != 'NA') & (data['gender'] != 'Unknown')]

data = data.dropna(subset=['age'])
data['mergedAgeGroup'] = data['age'].apply(AgeGroupMerge)

data['weekend'] = data['day'].apply(weekend)


data['naloxoneFirst'] = data['initialTreatment'].apply(lambda x: 1 if x.startswith('Naloxone') else 0)
```
